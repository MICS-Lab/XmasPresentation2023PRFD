```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(1)
```

```python
N = 100

plt.figure()
np.random.seed(1)
xs = np.random.rand(N)
ys = np.zeros_like(xs)
plt.scatter(xs, ys, c='r')
plt.grid(True)
plt.xlim(0, 1)
plt.ylim(-0.1, 0.1)
plt.yticks([0],[''])
# plt.title(f'{N} random points in 2D')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
np.random.seed(1)
xs = np.random.rand(N)
ys = np.random.rand(N)
zs = np.random.rand(N)
ax.scatter(xs, ys, zs, c='r')
plt.grid(True)
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_zlim(0, 1)
plt.title(f'{N} points in 3D')
plt.show()
```

```python
N = 100
D = 0.7
fig = plt.figure()
plt.grid(True)
np.random.seed(1)
xs = np.random.randn(N)-D
ys = np.random.randn(N)-D
plt.scatter(xs, ys, c='r')
np.random.seed(2)
xs = np.random.randn(N)+D
ys = np.random.randn(N)+D
plt.scatter(xs, ys, c='b')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.view_init(azim=-1, elev=0)
np.random.seed(1)
xs = np.random.randn(N)-D
ys = np.random.randn(N)-D
zs = np.random.randn(N)-D*2.5
ax.scatter(zs, xs, ys, c='r')
np.random.seed(2)
xs = np.random.randn(N)+D
ys = np.random.randn(N)+D
zs = np.random.randn(N)+D*2.5
ax.scatter(zs, xs, ys, c='b')
plt.tight_layout()
# plt.savefig('a.png', dpi=300)
plt.show()

```

```python
for a in [-20, -40, -60]:
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    ax.view_init(azim=a, elev=0)
    np.random.seed(1)
    xs = np.random.randn(N)-D
    ys = np.random.randn(N)-D
    zs = np.random.randn(N)-D*2.5
    ax.scatter(zs, xs, ys, c='r')
    np.random.seed(2)
    xs = np.random.randn(N)+D
    ys = np.random.randn(N)+D
    zs = np.random.randn(N)+D*2.5
    ax.scatter(zs, xs, ys, c='b')
    plt.tight_layout()
    # plt.savefig(f'a{a}.png', dpi=300)
    plt.show()
```

```python
def f(x,y):
    return 0.5*np.exp(-((x+1)**2+y**2)) + 0.02*((x-2)**2+(y)**2)
```

```python
plt.figure(figsize=(6,4))
xs = np.linspace(-5, 8, 200)
ys = f(xs,0)
plt.plot(xs, ys)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
xs = np.linspace(-5, 8, 200)
ys = f(xs,0)
plt.plot(xs, ys)
plt.xlabel('x')
plt.ylabel('f(x,0)')
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.view_init(azim=-65, elev=20)
X = np.arange(-5, 5, 0.25)
Y = np.arange(-4, 5, 0.25)
X, Y = np.meshgrid(X, Y)
# R = np.sqrt(X**2 + Y**2)
Z = f(X,Y)
surf = ax.plot_surface(X, Y, Z, cmap='Blues_r')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y)')
# Xs = np.linspace(-5, 5, 200)
# ax.plot(Xs, np.zeros_like(Xs), f(Xs,0), c='k', lw=1)
plt.tight_layout()
plt.savefig(f'b.png', dpi=300)
plt.show()
```

```python
N = 30
pis = np.zeros(N)
dims = np.arange(N)
pis[1] = 2
pis[2] = np.pi
for i in range(3,N):
    p = pis[i-2]
    pis[i] = p*(2*np.pi/i)

# print(pis)

plt.figure(figsize=(6,4))
plt.plot(dims[1:], pis[1:], '.')
plt.xlabel('$n$')
plt.ylabel('$n$-dimensional volume')
plt.grid(True)
plt.xlim(0, None)
plt.ylim(0, None)
plt.show()
```
