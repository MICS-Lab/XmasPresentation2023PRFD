```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(1)
```

```python
N = 100

plt.figure()
np.random.seed(1)
xs = np.random.rand(N)
ys = np.zeros_like(xs)
plt.scatter(xs, ys, c='r')
plt.grid(True)
plt.xlim(0, 1)
plt.ylim(-0.1, 0.1)
plt.yticks([0],[''])
# plt.title(f'{N} random points in 2D')
plt.show()
```

```python
plt.figure()
np.random.seed(1)
xs = np.random.rand(N)
ys = np.random.rand(N)
plt.scatter(xs, ys, c='r')
plt.grid(True)
plt.xlim(0, 1)
plt.ylim(0, 1)

# plt.title(f'{N} random points in 2D')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
np.random.seed(1)
xs = np.random.rand(N)
ys = np.random.rand(N)
zs = np.random.rand(N)
ax.scatter(xs, ys, zs, c='r')
plt.grid(True)
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_zlim(0, 1)
plt.title(f'{N} points in 3D')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
np.random.seed(1)
xs = np.random.rand(N)
ys = np.random.rand(N)
zs = np.zeros_like(xs)+0.5
ax.scatter(xs, ys, zs, c='r')
plt.grid(True)
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_zlim(0, 1)
plt.title(f'{N} points in 2D')
plt.show()
```
