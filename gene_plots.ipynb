```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(1)
```

```python
N = 100

plt.figure()
np.random.seed(1)
xs = np.random.rand(N)
ys = np.zeros_like(xs)
plt.scatter(xs, ys, c='r')
plt.grid(True)
plt.xlim(0, 1)
plt.ylim(-0.1, 0.1)
plt.yticks([0],[''])
# plt.title(f'{N} random points in 2D')
plt.show()
```

```python
plt.figure()
np.random.seed(1)
xs = np.random.rand(N)
ys = np.random.rand(N)
plt.scatter(xs, ys, c='r')
plt.grid(True)
plt.xlim(0, 1)
plt.ylim(0, 1)

# plt.title(f'{N} random points in 2D')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
np.random.seed(1)
xs = np.random.rand(N)
ys = np.random.rand(N)
zs = np.random.rand(N)
ax.scatter(xs, ys, zs, c='r')
plt.grid(True)
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_zlim(0, 1)
plt.title(f'{N} points in 3D')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
np.random.seed(1)
xs = np.random.rand(N)
ys = np.random.rand(N)
zs = np.zeros_like(xs)+0.5
ax.scatter(xs, ys, zs, c='r')
plt.grid(True)
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_zlim(0, 1)
plt.title(f'{N} points in 2D')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
np.random.seed(1)
xs = np.random.rand(N)
ys = np.zeros_like(xs)+0.5
zs = np.zeros_like(xs)+0.5
ax.scatter(xs, ys, zs, c='r')
plt.grid(True)
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_zlim(0, 1)
plt.title(f'{N} points in 1D')
plt.show()
```

```python
N = 11
plt.figure()
xs = np.arange(N)+1
ys = 2**xs
plt.plot(xs, ys, 'ro')
plt.ylim(-20, 1200)
plt.xlabel('$n$')
plt.ylabel('number of points')
plt.title('$n$-dimensional hypercube')
plt.show()
```

```python
N = 100
D = 0.7
fig = plt.figure()
plt.grid(True)
np.random.seed(1)
xs = np.random.randn(N)-D
ys = np.random.randn(N)-D
plt.scatter(xs, ys, c='r')
np.random.seed(2)
xs = np.random.randn(N)+D
ys = np.random.randn(N)+D
plt.scatter(xs, ys, c='b')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.view_init(azim=-1, elev=0)
np.random.seed(1)
xs = np.random.randn(N)-D
ys = np.random.randn(N)-D
zs = np.random.randn(N)-D*2.5
ax.scatter(zs, xs, ys, c='r')
np.random.seed(2)
xs = np.random.randn(N)+D
ys = np.random.randn(N)+D
zs = np.random.randn(N)+D*2.5
ax.scatter(zs, xs, ys, c='b')
plt.tight_layout()
# plt.savefig('a.png', dpi=300)
plt.show()

```

```python
for a in [-20, -40, -60]:
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    ax.view_init(azim=a, elev=0)
    np.random.seed(1)
    xs = np.random.randn(N)-D
    ys = np.random.randn(N)-D
    zs = np.random.randn(N)-D*2.5
    ax.scatter(zs, xs, ys, c='r')
    np.random.seed(2)
    xs = np.random.randn(N)+D
    ys = np.random.randn(N)+D
    zs = np.random.randn(N)+D*2.5
    ax.scatter(zs, xs, ys, c='b')
    plt.tight_layout()
    # plt.savefig(f'a{a}.png', dpi=300)
    plt.show()
```
